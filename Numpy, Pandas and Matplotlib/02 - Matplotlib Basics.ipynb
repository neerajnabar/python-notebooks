```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

## Generating series

```python
x = np.linspace(-1,1,101)
```

```python
y1 = np.sin(2*np.pi*x)
y2 = np.cos(2*np.pi*x)
y3 = y1/ y2
```

## Plotting functions

```python
plt.plot(x,y1,label='sine')
plt.plot(x,y2,'r',label='cosine')
plt.plot(x,y3,'g',label='tan')

plt.legend()   # enables legend, showing by the label param in plt.plot() function
plt.grid()     # enables grid
plt.show()     # shows the plot
```
